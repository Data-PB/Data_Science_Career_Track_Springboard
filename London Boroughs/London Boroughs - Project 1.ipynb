```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

```python
properties.head(2)
```

```python
# Transpose the dataframe
p_transpose = properties.transpose()
```

```python
# print few rows
p_transpose.head(2)
```

```python
# reseting index
p_transpose = p_transpose.reset_index()
```

```python
#Setting first row as header
p_transpose.columns = p_transpose.iloc[0]
```

```python
p_transpose.rename(columns = {'Unnamed: 0':'Boroughs_London', pd.NaT:'ID'}, inplace = True)
```

```python
p_transpose.head(2)
```

```python
#dropping row at index0 

p_transpose = p_transpose.drop(0)
```

```python
p_transpose.head(2)
```

```python
# too many columns, need to reconfigure the data with pd.melt
# Try this here: 
clean_df = pd.melt(p_transpose, id_vars= ['Boroughs_London', 'ID'],var_name ='Date',value_name = 'Price')
```

```python
clean_df
```

```python
# Are there more missing Boroughs in the dataset like Unnamed:47
clean_df.Boroughs_London.unique()
```

```python
# dropping the unnamed row as they don't have any data
cleaner_df = clean_df.dropna()
```

```python
cleaner_df
```

```python
# before getting into calculating average price make sure the data type of Price
cleaner_df.dtypes
```

```python
# change the data type for price to float
cleaner_df["Price"] = cleaner_df.Price.astype(float)
```

```python
# confirming the change 
cleaner_df.dtypes
```

```python
# A list of non-boroughs. Got from Tier 1 notebook 
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
```

```python
#filter dataframe to non-boroughs
cleaner_df[cleaner_df.Boroughs_London.isin(nonBoroughs)]
```

```python
cleaner_df = cleaner_df[~cleaner_df.Boroughs_London.isin(nonBoroughs)]
```

```python
cleaner_df
```

```python
# Visulize the dataset for one of the Boroughs - 'City of London'

city_london = cleaner_df[cleaner_df['Boroughs_London'] == 'City of London']
```

```python
plt.plot(city_london['Date'],city_london['Price'])
plt.xlabel('Date')
plt.ylabel('Year')
plt.title('Properties Price Flucation in City of London 1995-2020')
plt.show()
```

```python
#extract the year from date
cleaner_df['Year'] = cleaner_df['Date'].apply(lambda t: t.year)
```

```python
cleaner_df
```

```python
# groupby year and calculate average price for each year 

groupby_df = cleaner_df.groupby(['Boroughs_London', 'Year']).mean()
```

```python
groupby_df = groupby_df.reset_index()
```

```python
# function to calculate the average price 
def price_ratio(d):
    y95 = float(d['Price'] [d['Year']==1995])
    y21 = float(d['Price'] [d['Year']==2020])
    ratio = [y21/y95]
    return ratio


# def create_price_ratio(d):
#     y1998 = float(d['Average_price'][d['Year']==1998])
#     y2018 = float(d['Average_price'][d['Year']==2018])
#     ratio = [y2018/_ _ _]
#     return ratio
```

```python
# Making sure the function works, apllying it on one of the boroughs
price_ratio(groupby_df[groupby_df['Boroughs_London']=='Barking & Dagenham'])
```

```python
ratio = {}

for b in groupby_df['Boroughs_London'].unique():
    borough = groupby_df[groupby_df['Boroughs_London']== b]
    ratio[b] = price_ratio(borough)
print(ratio)    
    
```

```python
#save ratio as dataframe 

df = pd.DataFrame(ratio)
df
```

```python
# Transpose the df
df = df.transpose()
```

```python
df = df.reset_index()
```

```python
# lets rename the columns 
df = df.rename(columns={0:'Avg_Price_Increase','index':'Borough'})
```

```python
df.head()
```

```python
# sorting the datafarme in descending order
df = df.sort_values(by = ['Avg_Price_Increase'],ascending=False)
```

```python
#top10
df_10 = df.head(10)
```

```python
df_10
```

```python
#ploting the average price increase 
plt.figure(figsize=(15, 3))
plt.bar(df_10['Borough'],df_10['Avg_Price_Increase'],color='grey')
plt.title("Top 10 Boroughs with Highest Average Properties Price Increase in London")
plt.show()
```
